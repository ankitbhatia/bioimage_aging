```python
import numpy as np
from tqdm import tqdm,trange
from BioImage import BioImage, folders
import sys
from numpy.random import randint

data = np.load('filtered_dataset.npy')
rowOld = True
rowYoung = True
for row in tqdm(data):
    folder = folders[int(row[0])]
    index = int((row[1]))
    b = BioImage(folder, index)
    mask = b.getThresholded(b.ch1)
    ch2 = np.sum(np.multiply(mask, b.ch2))
    ch11 = np.sum(np.multiply(mask, b.ch11))
    p = np.array([ch2, ch11])
    if row[0]==0: # Old
        if rowOld: 
            points_old = p
            rowOld = False
        else:
            points_old = np.vstack((points_old, p))
    else: #Young
        if rowYoung:
            points_young = p
            rowYoung = False
        else:
            points_young = np.vstack((points_young,p))


   
    
```

```python
print(points_old.shape)
print(points_young.shape)
np.save('points_old', points_old)
np.save('points_young', points_young)
```
