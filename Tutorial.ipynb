```python
# check if the sample is an accept or a reject

b = BioImage(folders[0], 300, root='../Data/')
b.showImage()
extrema = b.getExtrema()
if (extrema[0] ==1 and extrema[1]==1):
    print('good data')
else:
    print('bad data')
    
```
